```python
# Étape 2 : Import des modules
import fitz  # PyMuPDF
from google.colab import files
import re
```

```python
# Étape 3 : Upload du fichier
uploaded = files.upload()
```

```python
# Étape 4 : Fonction d'extraction
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = ""
    for page in doc:
        full_text += page.get_text()
    doc.close()

    # Nettoyage : suppression de multiples espaces / sauts de ligne
    cleaned_text = re.sub(r'\n{2,}', '\n', full_text)
    cleaned_text = re.sub(r'[ \t]{2,}', ' ', cleaned_text)
    return cleaned_text.strip()
```

```python
# Étape 5 : Lancer l’extraction
for file_name in uploaded.keys():
    print(f"\n📄 Texte extrait de : {file_name}\n")
    texte_pdf = extract_text_from_pdf(file_name)
    print(texte_pdf[:3000])  # Affiche les 3000 premiers caractères
```
